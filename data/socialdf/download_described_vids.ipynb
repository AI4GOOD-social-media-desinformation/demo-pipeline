```python
import polars as pl
```

```python
df = pl.read_csv("socialdf_description.csv")
df.head()
```

```python
urls = df["url"].to_list()
urls
```

```python
import instaloader
import re

def download_instagram_video(target_folder, url):
    # Initialize Instaloader
    L = instaloader.Instaloader()

    # Matches /p/SHORTCODE or /reel/SHORTCODE
    shortcode_match = re.search(r'(?:/p/|/reel/)([^/?#&]+)', url)
    
    if not shortcode_match:
        raise ValueError("Could not parse shortcode from URL. Ensure URL contains /p/ or /reel/.")
    
    shortcode = shortcode_match.group(1)

    try:
        post = instaloader.Post.from_shortcode(L.context, shortcode)

        # Check if it is a video
        if not post.is_video:
            print("Target is not a video.")
            return

        print(f"Downloading video: {shortcode}...")
        L.download_post(post, target=shortcode)
        print("Download complete.")

    except Exception as e:
        print(f"Error: {e}")
```

```python
for v in urls:
    download_instagram_video(v)
```
