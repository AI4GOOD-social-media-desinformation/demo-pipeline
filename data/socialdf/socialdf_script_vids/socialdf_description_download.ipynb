# SocialDF Description Download Vids

```python
import polars as pl
from src.utils.download_insta_vids import download_insta_vids
```

```python
socialdf_desc = pl.read_csv("../socialdf_description.csv")
socialdf_desc.head()
```

```python
urls = socialdf_desc["url"].to_list()
for u in urls:
    download_insta_vids(u, output_folder="vids")
```

```python
import re
ids = []
urls = socialdf_desc["url"].to_list()

for u in urls:
    match = re.search(r"instagram\.com/p/([a-zA-Z0-9_-]+)", u)
    if match:
        ids.append(match.group(1))
    else:
        ids.append(None)
socialdf_desc = socialdf_desc.with_columns([pl.Series("id", ids)])
socialdf_desc.write_csv("processed_socialdf_description.csv")
```
